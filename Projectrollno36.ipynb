```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as p
auto=p.read_csv("Automobile_data.csv")
print(auto)

plt.figure(figsize=(10,6))
sns.lineplot(data=auto,x="length",y="width",hue="make",marker="o")
plt.title("Model Length VS Model Width")
plt.xlabel("Model Length (cm)")
plt.ylabel("Model Width (cm)")
plt.grid(True)
plt.show()
```

```python

auto['price'] = p.to_numeric(auto['price'], errors='coerce')
avg_length = auto.groupby("make")["price"].mean().reset_index()
plt.figure(figsize=(20, 5))
sns.barplot(data=avg_length, x="make", y="price", palette="magma")
plt.title("Average Price of Models")
plt.xlabel("Makes")
plt.ylabel("Avg Price")
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.histplot(data=auto,x="engine-size",hue="make",palette="dark",bins=15)
plt.title(" Distribution of Models by Engine size")
plt.xlabel("Engine size")
plt.ylabel("Frequency")

plt.show()
```
